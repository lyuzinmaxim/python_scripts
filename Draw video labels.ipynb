# Чтение картинок и лейблов в видео

```python
images_path = os.path.join("C:\\","Users","maksi","Videos","vis_first3","rus")
```

```python
os. chdir(images_path)
list_images = glob.glob('*.png')
list_labels = glob.glob('*.txt')
```

```python
list_images[99]
```

```python
for image in tqdm(list_images):
    img = cv2.imread(image)
 #   img = cv2.resize(img, (1024,1024), interpolation = cv2.INTER_AREA)

    #W, H,_ = img.shape
    W = 2052
    H = 2052
    #print(W,H)
    label = []
    
    with open(image[:-4]+'.txt', mode='r', encoding='utf-8') as f:
        label = [line.strip() for line in f]
        f.close()
    
    if label:
        for idx,el in enumerate(label):
            #x = np.zeros(5,)
            yolo_coords = np.fromstring(label[idx], sep=" ")
            
            x = np.array([yolo_coords[1]*W - W*yolo_coords[3]/2, 
                          yolo_coords[2]*H - H*yolo_coords[4]/2, 
                          yolo_coords[1]*W + W*yolo_coords[3]/2, 
                          yolo_coords[2]*H + H*yolo_coords[4]/2])
            
            cv2.rectangle(img, (int(x[0]),int(x[1])), (int(x[2]),int(x[3])), (0,0,255),2)
            #print("bbox: {} {} {} {}".format(int(x[0]),int(x[1]),int(x[2]),int(x[3])))

            cv2.putText(img,'БЛА{}'.format(idx+1),(int(x[0]),int(x[1])),cv2.FONT_HERSHEY_COMPLEX,0.7,(0,0,255),1,cv2.LINE_AA)
        cv2.imwrite(image[:-4]+'new'+'.png',img)
#        cv2.imshow('nn',img)
#        cv2.waitKey(0) 
```
