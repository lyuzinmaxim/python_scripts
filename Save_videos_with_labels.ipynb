```python
cap = cv2.VideoCapture(video_path)
print(cap)
```

```python
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
print('FPS:',fps)
print('Total frames:',total_frames)
```

```python
frame_list = np.arange(frame_start,frame_end+1,5)

print(frame_list,frame_list.shape);
```

```python
%%time
for i in range (frames_needed):
    cap.set(1,frame_list[i])
    ret, frame = cap.read()
    cur_time = np.ceil(frame_list[i] / fps);
    cur_time_min = int(cur_time//60);
    cur_time_sec = round(cur_time % 60);
    cv2.imwrite(os.path.join(saving_path , str(name_prefix) + str(frame_list[i]) + '_' + str(cur_time_min) + 'min' + str(cur_time_sec) + 'sec' + '.jpg'), frame)
    print(frame_list[i])
```

```python
import cv2
vidcap = cv2.VideoCapture('2.avi')
def getFrame(sec):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames,image = vidcap.read()
    if hasFrames:
        cv2.imwrite("image"+str(count)+".jpg", image)     # save frame as JPG file
    return hasFrames
sec = 0
frameRate = 3 #//it will capture image in each 0.5 second
count=1
success = getFrame(sec)
while success:
    count = count + 1
    sec = sec + frameRate
    sec = round(sec, 2)
    success = getFrame(sec)
```
