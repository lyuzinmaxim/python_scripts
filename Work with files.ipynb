```python
def change_class_in_file(path_to_file,list_of_files,class_needed):
    
    #path to folder with wiles should be like 'C:\\Users\\SaSha\\Documents\\Samples\\Labels'
    #class_needed is number of class, an integer
    #Добавить проверку на тип данных в файле!!!
    for i in range (len(list_of_files)):
                
        if os.stat(path_to_file + "\\" + list_of_files[i]).st_size == 0:
            print(path_to_file + "\\" + list_of_files[i],'is empty')
        
        else:
            a = np.genfromtxt(path_to_file + "\\" + list_of_files[i])
            a [0] = class_needed
    
            fmt = '%1.0f', '%1.6f', '%1.6f', '%1.6f', '%1.6f' 
            np.savetxt((path_to_file + "\\" + list_of_files[i]),a.reshape(1,5),newline="",fmt = fmt)
```

```python
def collect_data(path_to_file,list_of_files):
    
    #path to folder with wiles should be like 'C:\\Users\\SaSha\\Documents\\Samples\\Labels'
    #Добавить проверку на тип данных в файле!!!
    a = np.zeros(5)
    for i in range (len(list_of_files)):
                
        if os.stat(path_to_file + "\\" + list_of_files[i]).st_size == 0:
            print(path_to_file + "\\" + list_of_files[i],'is empty')
        
        else:
            b = np.genfromtxt(path + "\\" + listing[i])
            # a.append(i,axis = 0)
            a = np.vstack((a,b))

    a = np.delete(a,0,0)     
    return a
```
